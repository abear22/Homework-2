```python
import numpy as np
zscore=[]
main=[]
p=0

for x in range (0,lenx):
    for y in range (0,leny):
        m = t.iloc[[x],[y]]
        n = x_train.iloc[[x],[y]]
        p=p+1
        
        zscore.append(m)
        main.append(n)
        
k=np.array(zscore).reshape(lenx,leny)
l=np.array(main).reshape(lenx,leny)

out=[]
out= np.argwhere(abs(k>2))

print(df.median())

v1m=df['V1'].median()
v2m=df['V2'].median()
v3m=df['V3'].median()
v4m=df['V4'].median()
v5m=df['V5'].median()
v6m=df['V6'].median()
v7m=df['V7'].median()
v8m=df['V8'].median()
v8m=df['V9'].median()
```
